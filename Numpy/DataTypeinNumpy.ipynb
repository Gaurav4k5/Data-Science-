# changing data type

```python
import numpy as np
```

```python
arr = np.array([1.2, 2.4, 5.6])
print(arr.dtype)
```

```python
arr_int = arr.astype(np.int32)
print(arr_int)
print(arr_int.dtype)
```

# String  data type in numpy

```python
arr = np.array(['apple', 'banana', 'cherry'], dtype='U10') # unicode string array
print(arr)
```

# Complex Number

```python
arr= np.array([1+2j, 2+4j,2+6j], dtype='complex128')
print(arr)
```

```python
arr = np.array([{'a':1}, [1,3,4],'hello'], dtype= object)
print(arr)
```
