## Series - 1D Labeled Array

```python
import pandas as pd

s = pd.Series([10,20,30,40])
print(s)
```

```python
s = pd.Series([10,20,30], index=["a","b","c"])
```

```python
print(s)
```

## Dataframe - 2D Labeled Table

```python
data = {
    "name": ["Alice","Bob", "charlie"],
    "age": [25,30,35],
    "city": ["Delhi" ,"Mumbai", "Banglore"]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.index # Row labels
```

```python
df.columns
```

```python
df.index = ["a","b","c"]
df.columns = ["Name" , "Age" , "city"]
```

```python
print(df)
```
