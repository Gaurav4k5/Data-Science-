# Creating DataFrame

### 1 From python Lists

```python
import pandas as pd
data = [
["Alice", 25],
["Bob", 30],
["Charlie", 35]
]
df = pd.DataFrame(data, columns=["Name", "Age"])
(df)
```

### From Dictionary Of Lists

```python
data = {
    "Name" : ["Alice", "Bob", "Charlie"],
    "Age" : [25,64,44]
}
df = pd.DataFrame(data)
(df)
```

### From Numpy Arrays 

```python
import numpy as np

arr = np.array([[1,3],[4,5]])
df = pd.DataFrame(arr,columns = ["A", "B"])
df
```
