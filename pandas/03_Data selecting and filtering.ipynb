## Selecting columns

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
df['Actor']
```

```python
df[["Actor","Film"]]
```

```python
df.loc[4] # row by label
```

```python
df.iloc[4] # Row by position
```

```python
df.loc[0, "Actor"]        # Value at row 0, column 'Name'
```

```python
df.iloc[0, 1]            # Value at row 0, column at index 1
```

```python
df.loc[0:3, ["Actor" ,"Film", "IMDb"]]
```

```python
df.at[0, "Actor"]       # Fast label-based access
```

```python
df.iat[0, 1]           # Fast position-based access
```

## Filtering with Conditions

### Simple Condition

```python
df[df["IMDb"] > 8.0]
```

## Multiple Conditions (AND / OR)

```python
df[(df["Year"] > 2018) & (df["Genre"] == "Action/Drama")]
```

```python
df[(df["Year"] == 2018) | (df["IMDb"] < 5.0)]
```

```python
df.query("Year > 2018 and IMDb >= 7.0" )
```
