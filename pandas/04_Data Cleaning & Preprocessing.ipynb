# Handling Missing Values


```python
import pandas as pd
```

```python
df = pd.read_csv("data_cleaning_sample.csv")
```

```python
df
```

### Check for Missing Data

```python
 df.isnull()
```

```python
df.isnull().sum() # count missing per column
```

```python
df.dropna() 
```

```python
df.dropna(axis=1) # jis column me missing value hogi us column ko drop kar dega
```

```python
df.fillna(0)  # Replace NaN with 0
```

```python
df["Age"].fillna(df["Age"].mean()) # null ko replace kar dega mean se
```

```python
df.ffill()  # forword fill
```

```python
df.bfill() # backword fill
```

## Detecting & Removing Duplicates

```python
df.duplicated()  # true return karega duplicate pane per
```

```python
df.drop_duplicates()
```

```python
df.duplicated(subset=["Name", "Age"]) # perticuler jagah per search karega jaise Name Aur Age bola hia isme
```

```python
df["Email"].str.split("@")
```
