# Sorting & Ranking

### Sort by value

```python
df.sort_values("Year")    # Ascending sort
```

```python
df.sort_values("Year", ascending=False)  # decending
```

```python
df.sort_values(["Year", "IMDb"])
```

```python
df2 = df.sort_values(["Year", "IMDb"])
```

```python
df2
```

### Reset Index

```python
df2.reset_index(drop=True, inplace=True)
```

### sort by index

```python
df2.sort_index()
```
